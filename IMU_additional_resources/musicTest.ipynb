```python
import numpy as np
import scipy.io
import sounddevice as sd

# Load the .mat file
mat_data = scipy.io.loadmat('/Users/frank/Desktop/ENGN4213/Micro-Controller/IMU_additional_resources/IMU_additional_resources/IR_dataset/0_degree.mat')

# Extract the data from the mat_data dictionary
data = np.array(mat_data['ir'])

# Read data from text file
txt_data_1 = np.loadtxt('./data_1.txt')
txt_data_2 = np.loadtxt('./data_2.txt')

# Normalize the data
data_norm = data / np.max(np.abs(data))

# Normalize the text data
txt_data_norm_1 = txt_data_1 / np.max(np.abs(txt_data_1))
txt_data_norm_2 = txt_data_2 / np.max(np.abs(txt_data_2))

# Set the duration and sampling frequency
duration = 27  # Set the duration of the binaural signal (in seconds)
fs = 8000  # Set the sampling frequency (in Hz)

# Compute FFT of the signals
fft_left_channel = np.fft.fft(txt_data_norm_1)
fft_right_channel = np.fft.fft(txt_data_norm_2)
fft_hrtf_left = np.fft.fft(data_norm[:, 0], n=fft_left_channel.size) # Ensure that the FFTs are of same size
fft_hrtf_right = np.fft.fft(data_norm[:, 1], n=fft_right_channel.size)

# Multiply the frequency domain representations
binaural_left = fft_left_channel * fft_hrtf_left
binaural_right = fft_right_channel * fft_hrtf_right

# Compute the inverse FFT to obtain the time-domain signals
binaural_signal_time_1 = np.fft.ifft(binaural_left)
binaural_signal_time_2 = np.fft.ifft(binaural_right)

# The real part contains the binaural signal; the imaginary part should be negligible.
binaural_signal_time_1 = np.real(binaural_signal_time_1)
binaural_signal_time_2 = np.real(binaural_signal_time_2)

# Play the binaural signal
# sd.play(binaural_signal_time_1, fs)
# sd.wait()

# Save the binaural signal to a text file
np.savetxt('binaural_signal_1.txt', binaural_signal_time_1, delimiter='\t')
np.savetxt('binaural_signal_2.txt', binaural_signal_time_2, delimiter='\t')
```
